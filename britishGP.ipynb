<h1>F1 ANALYSIS</h1>

```python
import fastf1 as f1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1.Cache.enable_cache('cache')
```

```python
race = f1.get_session(2022,'British','R')
race.load()
```

```python
ver_laps = race.laps.pick_driver("VER")
ver_sector1 = ver_laps['SpeedI1'].dropna().astype(int)
lec_laps = race.laps.pick_driver("LEC")
lec_sector1 = lec_laps['SpeedI1'].dropna().astype(int)
```

```python
plt.plot(ver_sector1,label="Ver")
plt.plot(lec_sector1)
plt.show()

```
