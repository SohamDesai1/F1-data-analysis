```python
import fastf1 as f1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1.Cache.enable_cache('cache')
```

```python
race = f1.get_session(2023,'Saudi Arabia','FP1')
race.load()
```

```python
race = f1.get_session(2023,'Saudi Arabia','R')
race.load()
```

```python
race.results.loc[:, ['DriverNumber', 'Abbreviation', 'TeamName', 'Position', 'Q1', 'Q3', 'Q2']]
```

```python
laps_max = race.laps.pick_driver('VER').pick_fastest()
telemetry_max = laps_max.get_telemetry().add_distance()
speed_max = telemetry_max.Speed

laps_alo = race.laps.pick_driver('ALO').pick_fastest()
telemetry_alo = laps_alo.get_telemetry().add_distance()
speed_alo = telemetry_alo.Speed
```

```python
# plot two fastest laps
plt.plot(telemetry_max.Distance, speed_max, label='Verstappen')
plt.plot(telemetry_alo.Distance, speed_alo, label='Alomnso')
plt.legend()
plt.show()
```

```python
laps_rus = race.laps.pick_driver('RUS').pick_fastest()
telemetry_rus = laps_rus.get_telemetry().add_distance()
speed_rus = telemetry_rus.Speed
```

```python
plt.plot(telemetry_max.Distance, speed_max, label='Verstappen')
plt.plot(telemetry_rus.Distance, speed_rus, label='Russel')
plt.legend()
plt.show()
```
