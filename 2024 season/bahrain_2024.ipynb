# Bahrain Grand Prix 2024

**Qualifying**

```python
quali = f1.get_session(2024,"Bahrain","Qualifying")
quali.load()
```

```python
quali.session_info
```

```python
quali.results
```

```python
ver = quali.get_driver("VER")
ver
```

```python
laps_max = quali.laps.pick_driver("VER")
laps_max.loc[:,["LapTime","Sector1Time","Sector2Time","Sector3Time","SpeedI1","SpeedI2","SpeedFL","SpeedST"]]
```

```python
fastest_lap_max = quali.laps.pick_driver("VER").pick_fastest()
tele_max= fastest_lap_max.get_telemetry().add_distance()
speed_max = tele_max.Speed
tele_max
```

```python
laps_lec = quali.laps.pick_driver("LEC").pick_fastest()
tele_lec = laps_lec.get_telemetry().add_distance()
speed_lec = tele_lec.Speed
tele_lec
```

```python
laps_alo = quali.laps.pick_driver("ALO").pick_fastest()
tele_alo = laps_alo.get_telemetry().add_distance()
speed_alo = tele_alo.Speed
tele_alo
```

```python
plt.plot(tele_max.Distance,speed_max,label="Verstappen")
plt.plot(tele_alo.Distance,speed_alo,label = "Alonso")
plt.plot(tele_lec.Distance,speed_lec,label = "Leclerc")
plt.legend()
plt.show()
```

**Race**

```python
race = f1.get_session(2024,"Bahrain","R")
race.load()
```

```python
race.results.loc[:,["Abbreviation","TeamName","GridPosition","Position","Time","Status"]]
```

```python
ver_race = race.laps.pick_driver("VER")
fastest_lap_max_race = ver_race.pick_fastest().LapTime
print("Fastest Lap:",fastest_lap_max_race,"which was in Lap number",ver_race.pick_fastest().LapNumber)
time_columns = ["LapTime", "Sector1Time", "Sector2Time", "Sector3Time"]
ver_race.loc[:, time_columns] = ver_race.loc[:, time_columns].apply(pd.to_timedelta)
ver_race.loc[:,["LapNumber","LapTime","Sector1Time","Sector2Time","Sector3Time","SpeedI1","SpeedI2"]]
```

```python
lec_race = race.laps.pick_driver("LEC")
fastest_lap_lec_race = lec_race.pick_fastest().LapTime
print("Fastest Lap:",fastest_lap_lec_race,"which was in Lap number",lec_race.pick_fastest().LapNumber)
time_columns = ["LapTime", "Sector1Time", "Sector2Time", "Sector3Time"]
lec_race.loc[:, time_columns] = lec_race.loc[:, time_columns].apply(pd.to_timedelta)
lec_race.loc[:,["LapNumber","LapTime","Sector1Time","Sector2Time","Sector3Time","SpeedI1","SpeedI2"]]
```

```python
fastest_lap_race_lec = race.laps.pick_driver("LEC").pick_fastest()
tele_race_lec = fastest_lap_race_lec.get_telemetry().add_distance()
speed_race_lec = tele_race_lec.Speed
tele_race_lec
```

```python
fastest_lap_race_max = race.laps.pick_driver("VER").pick_fastest()
tele_race_max= fastest_lap_race_max.get_telemetry().add_distance()
speed_race_max = tele_race_max.Speed
tele_race_max
```

```python
plt.plot(tele_race_max.Distance,speed_race_max,label="Verstappen")
plt.plot(tele_race_lec.Distance,speed_race_lec,label = "Leclerc")
plt.legend()
plt.show()
```
