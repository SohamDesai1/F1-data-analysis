```python
import fastf1 as f1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1.Cache.enable_cache('cache')
```

```python
race = f1.get_session(2020, 'Turkish Grand Prix', 'Q')
race.load()
```

```python
race.results.loc[:, ['DriverNumber', 'Abbreviation', 'TeamName', 'Position', 'Q1', 'Q3', 'Q2']]
```

```python
race.laps
```

```python
laps_max = race.laps.pick_driver('VER').pick_fastest()
telemetry_max = laps_max.get_telemetry().add_distance()
speed_max = telemetry_max.Speed
```

```python
laps_lec = race.laps.pick_driver('LEC').pick_fastest()
telemetry_lec = laps_lec.get_telemetry().add_distance()
speed_lec = telemetry_lec.Speed
```

```python
laps_lec = race.laps.pick_driver('LEC').pick_fastest()
telemetry_lec = laps_max.get_telemetry().add_distance()
speed_lec = telemetry_lec.Speed
```

```python
# plot two fastest laps
plt.plot(telemetry_max.Distance, speed_max, label='Verstappen')
plt.plot(telemetry_lec.Distance, speed_lec, label='Leclerc')
plt.legend()
plt.show()
```

```python
race2 = f1.get_session(2022,'Australia','R')
race2.load()
```

```python
race2.results
```

```python
laps_lec1 = race2.laps.pick_driver('LEC').pick_fastest()
telemetry_lec1 = laps_lec1.get_telemetry().add_distance()
speed_lec1 = telemetry_lec1.Speed

laps_per = race2.laps.pick_driver('PER').pick_fastest()
telemetry_per = laps_per.get_telemetry().add_distance()
speed_per = telemetry_per.Speed

# plot two fastest laps
plt.plot(telemetry_lec1.Distance, speed_lec1, label='Leclerc')
plt.plot(telemetry_per.Distance, speed_per, label='Perez')
plt.legend()
plt.show()
```
